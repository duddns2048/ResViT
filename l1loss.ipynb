```python
import torch
import torch.nn as nn

# 예측 값 (predictions)와 실제 값 (targets)
predictions = torch.tensor([[2.5, 0.0, 2.1, 0.0], [1.0, 3.2, 1.5,0.5]], requires_grad=True)
targets = torch.tensor([[3.0, -0.5, 2.0, 0.0], [1.0, 2.0, 1.0, 0.0]])


# predictions = torch.tensor([2.0], requires_grad=True)
# targets = torch.tensor([0.0])


# L1Loss 함수 정의
criterion = nn.L1Loss()

# 손실 계산
loss = criterion(predictions, targets)

print('L1 Loss:', loss.item())

# 역전파를 통해 그래디언트를 계산
loss.backward()

# 예측 값에 대한 그래디언트 출력
print('Gradients:', predictions.grad)
```

```python
import visdom
import numpy as np

# python -m visdom.server -port 8097

# Visdom 인스턴스 생성
vis = visdom.Visdom()

# 환경 설정 (여기서는 'main'이라는 환경 사용)
env = 'main'
vis = visdom.Visdom(env=env)

# 예제 데이터 플롯
x = np.linspace(0, 10, 100)
y = np.cos(x)
vis.line(X=x, Y=y, opts=dict(title='Sine Wave', xlabel='x', ylabel='y'))
```
