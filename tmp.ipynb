```python
import os

# 파일이 있는 디렉터리 경로 지정
directory = './results/fold0/'

# 디렉터리 내 모든 파일 및 디렉터리 목록 가져오기
for filename in os.listdir(directory):
    # 파일 경로 생성
    file_path = os.path.join(directory, filename)
    
    # 디렉터리인 경우 스킵
    if os.path.isdir(file_path):
        continue
    
    # 새로운 파일 이름 생성
    new_filename = 'BraTS-GLI-' + filename
    
    # 새로운 파일 경로 생성
    new_file_path = os.path.join(directory, new_filename)
    
    # 파일 이름 변경
    os.rename(file_path, new_file_path)

print("모든 파일의 이름 변경 완료")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값의 범위를 설정합니다. log(x)는 x > 0인 경우에만 정의됩니다.
x = np.linspace(0, 1, 400)

# -log(1-x)를 계산합니다.
y = -np.log(x)

# 그래프를 그립니다.
plt.plot(x, y, label='-log(x)')

# 그래프의 제목과 축 레이블을 추가합니다.
plt.title('-log(x) Graph')
plt.xlabel('x')
plt.ylabel('-log(x)')

# 범례를 추가합니다.
plt.legend()

# y=0 축을 더 굵게 그립니다.
plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=1)
plt.axvline(1, color='black', linewidth=1)

# x축 눈금을 0.2 단위로 설정합니다.
plt.xticks(np.arange(0, 1.1, step=0.2))

# 격자 선의 두께를 조정합니다.
plt.grid(True, linewidth=0.5)

# 그래프를 표시합니다.
plt.show()
```
